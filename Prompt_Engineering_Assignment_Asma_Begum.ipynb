# Assignment: Prompt Engineering for Performance Improvement
**Author:** Asma Begum

## Constant Input
```
So I grabbed this “XJ200 blender” last week on Amazon on July 8th — think it was around 79 dollars but not sure if it was on sale.
Anyway, Maria L. here. Honestly, kinda disappointed. Works OK but the motor smells funny sometimes. I'd rate it 3 out of 5.
```

## Baseline Prompt & Output
### Prompt:
```
Extract the data.
```
### Output:
The reviewer bought a blender and didn’t like the motor smell.

## Technique 1 — Role Prompting
### Prompt:
```
You are a Senior Data Extraction Analyst. Extract the key details: product, price, reviewer, date, sentiment, summary.
```
### Output:
Product: XJ200 blender

Price: around 79 dollars

Reviewer: Maria L.

Date: July 8th

Sentiment: negative

Summary: Reviewer is disappointed because the motor smells odd.

## Technique 2 — JSON Formatting
### Prompt:
```
Extract the data and output ONLY valid JSON:
{
  "product": "",
  "price": "",
  "reviewer": "",
  "date": "",
  "sentiment": "",
  "summary": ""
}
```
### Output:
```
{
  "product": "XJ200 blender",
  "price": "79 dollars",
  "reviewer": "Maria L.",
  "date": "July 8th",
  "sentiment": "negative",
  "summary": "The reviewer is disappointed because the motor smells strange."
}
```

## Technique 3 — Chain-of-Thought Reasoning
### Prompt:
```
Think step-by-step about the entities, then output ONLY valid JSON.
```
### Output:
```
{
  "product": "XJ200 blender",
  "price": "79 dollars",
  "reviewer": "Maria L.",
  "date": "July 8th",
  "sentiment": "negative",
  "summary": "The reviewer purchased the blender for about $79 and was disappointed because the motor smells strange."
}
```

## Final Optimized Prompt & Output
### Prompt:
```
You are a Senior Data Extraction Analyst with expertise in ambiguity resolution and strict JSON formatting.
Follow steps and output only valid JSON.
```
### Output:
```
{
  "product": "XJ200 blender",
  "price": "79 dollars",
  "reviewer": "Maria L.",
  "date": "July 8th",
  "sentiment": "negative",
  "summary": "Maria purchased the XJ200 blender for about $79 on July 8th and was disappointed because the motor emits a strange odor."
}
```
